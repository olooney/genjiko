```python
g4 = list(generate_all_genjiko_patterns(n=4))

def sort_key(genjiko: GenjikoType):
    partition = [ p for h, p in genjiko ]
    sizes = [ (min(group), len(group)) for group in partition ]
    return [ size for left, size in sorted(sizes) ]

sorted(g4, key=sort_key)
```

```python
draw_genjiko_grid(
    genjiko_patterns=sorted(generate_all_genjiko_patterns(n=1), key=sort_key),
    grid_width=1,
    grid_height=1,
    rubricate=True,
)
```
